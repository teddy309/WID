```python
from kafka import KafkaConsumer
from json import loads



# kafka consumer connect
consumer = KafkaConsumer(
    'package1','package2',
    bootstrap_servers=['localhost:9092'],
    auto_offset_reset='earliest',
    enable_auto_commit=True,
    group_id='groupx',
    value_deserializer=lambda x: loads(x.decode('utf-8')))





#x, y = [], []
last_x1 = -100
last_y1 = -100
last_x2 = -100
last_y2 = -100

stream1 = []
stream2 = []
# receive data from kafka producer and draw real time graph
for message in consumer:
    message = message.value
    
    if list(message.keys())[0] == 'data1':
        print('Data 1 exist')
        if len(stream1) == 5:
            stream1.pop(0)
        else:
            stream1.append(message)
        last_x1 = message['x']
        last_y1 = message['y']
    if list(message.keys())[0] == 'data2':
        print('Data 2 exist')
        if len(stream2) == 5:
            stream1.pop(0)
        else:
            stream1.append(message)

        last_x2 = message['x']
        last_y2 = message['y']
    
    if (last_x1 != -100 and last_y1 != -100 and last_x2 != -100 and last_y2 != -100):
        if (last_x1 == last_x2) and (last_y1 == last_y2):
            print('CONFLICT')
        else:
            print('NOT CONFLICT')
    
    print(message)
```
